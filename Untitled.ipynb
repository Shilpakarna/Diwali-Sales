EXPLORATORY DATA ANALYSIS


1. ON THE BASIS OF GENDER

```python
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

2. On the basis of Age

```python
ax=sns.countplot(x='Age Group', data=df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Gender')
```

```python
df.columns
```

3. On the basis of Marital Status

```python
ax=sns.countplot(x='Marital_Status', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_married=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=sales_married)
```

4. On the basis of State

```python
ax=sns.countplot(x='State', data=df)

sns.set(rc={'figure.figsize': (15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='State',y='Amount',data=sales_state)
```

5. On the basis of Occupation

```python
ax=sns.countplot(x='Occupation', data=df)

sns.set(rc={'figure.figsize': (15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales_occu)
```

6. On the basis of Product Category

```python
ax=sns.countplot(x='Product_Category', data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_product)

sns.set(rc={'figure.figsize':(20,3)})
```
